# **Validação**

```python
schema = pa.DataFrameSchema(
    columns = {
        'city': pa.Column(pa.String),
        'city_ascii': pa.Column(pa.String),
        'lat': pa.Column(pa.Float),
        'lng': pa.Column(pa.Float),
        'country': pa.Column(pa.String),
        'iso': pa.Column(pa.String, pa.Check.str_length(3,3)),
        'capital': pa.Column(pa.String, nullable = True),
        'population': pa.Column(pa.Float, nullable = True)
    }
)
```

```python
schema.validate(df)
```

*Pela validação, percebe-se que existem valores ausentes para as colunas:*

*'capital' e 'population'*


```python
df.isna().sum()
```

*Caso queiramos retirar todas as linhas que contém pelo menos um valor ausente:*

```python
df.dropna()
```

*Localizar as linhas com valores ausentes*

```python
df.loc[df.capital.isnull() | df.population.isnull()]
```

*Retirar todas as linhas que contém pelo menos um valor ausente mas especificando a coluna requerida:*

```python
df.dropna(subset = ['capital'])
```

*Criando um backup pra coluna 'capital' já que existem muitos valores ausentes e iremos querer possivelmente adaptar esses valores>*

```python
df['capital_backup'] = df.capital
df
```

```python
df.drop(labels = ['capital_backup'], axis = 1, inplace = True)
```

*Retirar linhas duplicadas*

```python
df.drop_duplicates()
```

*Logo, não existem linhas duplicadas.*

# **Etapa de Transformação**

```python
filtro_country = df.country == 'Japan'
filtro_capital1 = df.capital == 'admin'
filtro_capital2 = df.capital == 'primary'
filtro_capital = filtro_capital1 | filtro_capital2
filtro_population = df.population > 1000000

df.loc[filtro_country & filtro_capital & filtro_population]
```
